```python
import pandas
import cvxpy as cp
import numpy as np

# Load data 
my_input =  "https://basket.co.il/pbp/json/players-stars.json"
with urllib.request.urlopen(my_input) as url:
    mydata = json.loads(url.read().decode('utf-8-sig'))
mydata = pd.DataFrame(mydata[0]['players'])
weights = mydata['cost'].values
is_play = mydata['playing_next'].values
teams = mydata['team_id'].values
team_count = mydata.team_id.nunique()

values = mydata['average_pir'].values
players_count = len(weights)
# get players to team matrix
temp = mydata[['PlayerNameHeb','team_name']]
one_hot = pd.get_dummies(temp,columns=['team_name'])
one_hot.drop(columns=['PlayerNameHeb'],inplace=True)

# The variable we are solving for: a binary vector whose length
# is given by the total number of 
selection = cp.Variable(players_count, boolean=True)

# The total cost should be no more than 100, and the team
# cannot contain more than 5 players
cost_constraint = weights * selection <= 100
size_constraint = cp.sum(selection) <= 5
is_playing_constraint = selection <= is_play
team_price_constraint = cp.multiply(selection, weights).T * one_hot <= np.ones(team_count) * 100
# Our total utility is the sum of the item utilities
total_utility = values * selection

# We tell cvxpy that we want to maximize total utility 
# subject to cost_constraint and size_constraint
knapsack_problem = cp.Problem(
    cp.Maximize(total_utility),
    [cost_constraint, size_constraint,is_playing_constraint,team_price_constraint]
)

# Solve the problem
best_score = knapsack_problem.solve()
print(knapsack_problem.status)
# Retrieve  indices
indices = [i for i, val in enumerate(selection.value) if val >= 0.9] 

# # Display the dream team and its score
print(mydata.ix[indices])
print(best_score)
```
